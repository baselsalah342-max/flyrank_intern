## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

The Search Question: Which individual content pages are showing evidence of traffic decay or holding untapped visibility potential, and should be prioritized for a refresh review?

Unit of Analysis: A single content item per client (content_id / content_hash_id).

The Decision & Action: This work improves the editorial team's scheduling decision. Instead of guessing which pages to update, an editor will act directly on our ranked opportunity queue, reviewing the top-K pages alongside their specific reason codes to update, expand, or protect them.

The Cost of a Wrong Call: A false positive (recommending a healthy, stable page for a refresh) wastes limited human editorial hours on unnecessary rewrites. Even worse, it risks breaking the page's current successful structure, causing an accidental drop in its stable search engine ranking.

Why ML/Data helps: With thousands of content items across multiple clients, manually tracking continuous trends, keyword-mix decay, and engagement drops simultaneously is impossible for human teams. Machine learning and data pipelines can systematically calculate multi-signal risk profiles to rank these opportunities at scale.
